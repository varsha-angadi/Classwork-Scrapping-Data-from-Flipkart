```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.flipkart.com/search?q=Laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off')
```

```python
soup=BeautifulSoup(driver.page_source,'html.parser')
```

```python
soup
```

```python
name='KzDlHZ'
actual_price='yRaY8j ZYYwLA'
offered_price='Nx9bqj _4b5DiR'
```

```python
Laptop_name=soup.find_all('div',class_=name)
Actual_price=soup.find_all('div',class_=actual_price)
Offered_price=soup.find_all('div',class_=offered_price)
```

```python
name_list=[]
for i in Laptop_name:
    name_list.append(i.text)
name_list
```

```python
actual_price_list=[]
for i in Actual_price:
    actual_price_list.append(i.text)
actual_price_list
```

```python
offered_price_list=[]
for i in Offered_price:
    offered_price_list.append(i.text)
offered_price_list
```

```python
df = pd.DataFrame({
        'Laptop': name_list,
        'Actual Price': actual_price_list,
        'Offered Price': offered_price_list
    })
```

```python
df
```

```python
df.to_csv('flipkart_laptops.csv', index=False)
```

```python
print("Scraping completed and data saved to flipkart_laptops.csv")
```
